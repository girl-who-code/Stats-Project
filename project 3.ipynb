```python
# Univariate analysis


# Bivariate analysis 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Load your data (replace this with your actual data)
data = {
    'Sales': [500, 600, 700, 800, 550, 650, 750, 850, 900, 1000],
    'Profit': [50, 80, 90, 120, 60, 100, 110, 140, 130, 160],
    'Quantity': [5, 6, 7, 8, 5, 6, 7, 8, 9, 10]
}
df = pd.DataFrame(data)

# --- Univariate Analysis ---

# 1. Summary Statistics
print("\nSummary Statistics for Sales, Profit, and Quantity:")
print(df.describe())

# 2. Histogram for Sales
plt.figure(figsize=(8, 6))
sns.histplot(df['Sales'], kde=True, color='blue', bins=10)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()

# 3. Box Plot for Profit
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Profit'], color='green')
plt.title('Box Plot of Profit')
plt.xlabel('Profit')
plt.show()

# --- Bivariate Analysis ---

# 1. Scatter Plot between Sales and Profit
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Sales', y='Profit', data=df, color='purple')
plt.title('Scatter Plot between Sales and Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()

# 2. Correlation Heatmap between variables
corr_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# 3. Pairplot for Sales, Profit, and Quantity
sns.pairplot(df, hue='Quantity', palette='coolwarm')
plt.suptitle('Pairplot: Sales, Profit, and Quantity', y=1.02)
plt.show()
```

```python

# Plotting sales trends over time



# Resample sales data by month

import pandas as pd
import matplotlib.pyplot as plt

# Example: Load your data (replace this with your actual data)
data = {
    'Date': ['2023-01-01', '2023-01-15', '2023-02-01', '2023-02-15', 
             '2023-03-01', '2023-03-15', '2023-04-01', '2023-04-15'],
    'Sales': [200, 220, 250, 270, 300, 310, 330, 350]
}
df = pd.DataFrame(data)

# Convert the Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Set the Date column as the index
df.set_index('Date', inplace=True)

# Resample data by month and sum the sales
monthly_sales = df['Sales'].resample('M').sum()

# Plot the sales trends over time
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='b')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
# Plotting profit trends over time


import pandas as pd
import matplotlib.pyplot as plt

# Example: Load your data (replace this with your actual data)
data = {
    'Date': ['2023-01-01', '2023-01-15', '2023-02-01', '2023-02-15', 
             '2023-03-01', '2023-03-15', '2023-04-01', '2023-04-15'],
    'Profit': [500, 600, 700, 750, 800, 850, 900, 950]
}
df = pd.DataFrame(data)

# Convert the Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Set the Date column as the index
df.set_index('Date', inplace=True)

# Resample data by month and sum the profits
monthly_profit = df['Profit'].resample('M').sum()

# Plot the profit trends over time
plt.figure(figsize=(10, 6))
plt.plot(monthly_profit.index, monthly_profit.values, marker='o', linestyle='-', color='g')
plt.title('Monthly Profit Trends')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python

# Identifying top-performing product categories based on total sales and profit

# Plotting total sales by product category

# Plotting total profit by product category

import pandas as pd
import matplotlib.pyplot as plt

# Example: Load your data (replace this with your actual data)
data = {
    'Product Category': ['Electronics', 'Furniture', 'Electronics', 'Clothing', 
                         'Clothing', 'Furniture', 'Electronics', 'Clothing'],
    'Sales': [1000, 1500, 2000, 500, 700, 1200, 1800, 600],
    'Profit': [200, 300, 400, 100, 150, 250, 350, 120]
}
df = pd.DataFrame(data)

# Group by 'Product Category' and calculate total Sales and Profit
category_performance = df.groupby('Product Category').agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).sort_values(by='Sales', ascending=False)

# Display top-performing product categories
print(category_performance)

# Plot the top categories based on Sales and Profit
plt.figure(figsize=(12, 6))
category_performance.plot(kind='bar', figsize=(12, 6), color=['blue', 'green'], alpha=0.8)
plt.title('Top-Performing Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Total Amount')
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.show()


```

```python
# Visualizing the correlation between sales and profit

# Displaying the correlation coefficient

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example: Load your data (replace this with your actual data)
data = {
    'Sales': [200, 300, 150, 400, 250, 320, 180, 420],
    'Profit': [50, 80, 30, 120, 60, 90, 40, 130]
}
df = pd.DataFrame(data)

# Calculate the correlation coefficient
correlation = df['Sales'].corr(df['Profit'])
print(f"Correlation Coefficient between Sales and Profit: {correlation:.2f}")

# Visualize the correlation using a scatter plot with a regression line
plt.figure(figsize=(8, 6))
sns.regplot(x='Sales', y='Profit', data=df, color='b', line_kws={"color": "red"})
plt.title('Correlation between Sales and Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example: Load your data (replace this with your actual data)
data = {
    'Date': ['2021-01-15', '2021-02-20', '2021-03-25', '2021-04-10', 
             '2022-01-18', '2022-02-15', '2022-03-10', '2022-04-20',
             '2023-01-22', '2023-02-12', '2023-03-18', '2023-04-28'],
    'Sales': [500, 600, 700, 800, 550, 650, 750, 850, 600, 700, 800, 900]
}
df = pd.DataFrame(data)

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract Year and Month from Date
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

# Group by Year and Month and sum Sales
seasonal_sales = df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()

# Create a pivot table for Year (rows) and Month (columns)
sales_matrix = seasonal_sales.pivot('Year', 'Month', 'Sales')

# Plot a heatmap to visualize seasonal sales trends
plt.figure(figsize=(12, 8))
sns.heatmap(sales_matrix, annot=True, fmt='.0f', cmap='coolwarm', cbar_kws={'label': 'Total Sales'})
plt.title('Seasonal Sales Trends (Year vs. Month)')
plt.xlabel('Month')
plt.ylabel('Year')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example: Load your data (replace this with your actual data)
data = {
    ' Ship Date ': ['2021-01-15', '2021-02-20', '2021-03-25', '2021-04-10', 
                    '2022-01-18', '2022-02-15', '2022-03-10', '2022-04-20'],
    'Sales': [500, 600, 700, 800, 550, 650, 750, 850]
}
df = pd.DataFrame(data)

# Clean column names
df.columns = df.columns.str.strip()          # Remove leading/trailing spaces
df.columns = df.columns.str.replace(' ', '_') # Replace spaces with underscores
df.columns = df.columns.str.lower()          # Convert to lowercase

# Print cleaned column names
print(f"Cleaned Column Names: {df.columns}")

# Ensure 'Ship Date' column is datetime type
df['ship_date'] = pd.to_datetime(df['ship_date'], errors='coerce')

# Drop rows where 'Ship Date' couldn't be converted to datetime
df = df.dropna(subset=['ship_date'])

# Resample the sales data by month and sum the sales
df.set_index('ship_date', inplace=True)
monthly_sales = df['sales'].resample('M').sum()

# Plot the resampled sales data by month
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker='o', color='b')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```
